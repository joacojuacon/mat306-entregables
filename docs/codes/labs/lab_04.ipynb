````python
# una función auxiliar
def resumen_df(df):
    """
    Función para generar un resumen de un DataFrame que incluye
    el número de elementos distintos y vacíos por columna.

    Args:
    df (pd.DataFrame): DataFrame a resumir.

    Returns:
    pd.DataFrame: DataFrame resumen con el nombre de las columnas,
                  cantidad de elementos distintos y cantidad de elementos vacíos.
    """

    nombres = df.columns
    result = pd.DataFrame({'nombres': nombres})

    result['elementos_distintos'] = df.nunique().values

    result['elementos_vacios'] = df.isnull().sum().values

    result['porcentaje_elementos_vacios'] = df.isnull().sum().values/len(df)*100


    return result
````

En esta sección se busca identificar cuáles son los países de América Latina que han presentado los valores extremos del **Índice de Libertad de Prensa** en cada año observado.

> Recuerda que un menor puntaje en `indice` implica mayor libertad de prensa.

#### **Tareas:**

**a)** Utilizando un ciclo `for`, recorre cada año del conjunto de datos filtrado por países latinoamericanos, y determina para cada año:

* El país con el menor valor de `indice` (mayor libertad de prensa).
* El país con el mayor valor de `indice` (menor libertad de prensa).

**b)** Resuelve la misma tarea del punto anterior utilizando un enfoque vectorizado con `groupby`, sin usar ciclos explícitos.



#### **Lista de países latinoamericanos considerada:**

```python
america = ['ARG', 'ATG', 'BLZ', 'BOL', 'BRA', 'CAN', 'CHL', 'COL', 'CRI',
```

> Puedes usar esta lista para filtrar el DataFrame final por la columna `codigo_iso`.



````python
# respuesta
america = ['ARG', 'ATG', 'BLZ', 'BOL', 'BRA', 'CAN', 'CHL', 'COL', 'CRI',
       'CUB', 'DOM', 'ECU', 'GRD', 'GTM', 'GUY', 'HND', 'HTI', 'JAM',
       'MEX', 'NIC', 'PAN', 'PER', 'PRY', 'SLV', 'SUR', 'TTO', 'URY',
       'USA', 'VEN']

#a) Meidante un ciclo for
anio = df['anio'].unique()

for a in anio:
       df_anio = df[df['anio'] == a]
       df_anio_america = df_anio[df_anio['codigo_iso'].isin(america)]
       df_anio_america_mayormenor = df_anio_america.sort_values(by='indice',ascending = False).fillna(0)
       df_anio_america_menormayor = df_anio_america.sort_values(by='indice',ascending = True).fillna(0)

       print('En el año {0} el país con mayor índice fue {1}, el cual fue {2}'.format(a,df_anio_america_mayormenor.iloc[0]['pais'],df_anio_america_mayormenor.iloc[0]['indice']))
       print('En el año {0} el país con menor índice fue {1}, el cual fue {2}'.format(a,df_anio_america_menormayor.iloc[0]['pais'],df_anio_america_menormayor.iloc[0]['indice']))
````

````python
#b) Mediante un Groupby

#Para el máximo
df_america = df[df['codigo_iso'].isin(america)].dropna()
df_americaMaximoIndiceAnio = df_america.groupby('anio')['indice'].max().fillna(0).reset_index()

df_americaMaximo = pd.merge(df_america,df_americaMaximoIndiceAnio,on=['anio','indice'],how = 'inner')
df_americaMaximo
````
